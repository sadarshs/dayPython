NUMPY

```python
import pandas as pd
import numpy as np

from google.colab import files
uploaded = files.upload()

df = pd.read_csv('1774404110.csv')

df.head()
```

```python
mpg = df['mpg'].to_numpy()

print("Mean:", np.mean(mpg))
print("Median:", np.median(mpg))
print("Std Dev:", np.std(mpg))
print("MPG > 25:", np.sum(mpg > 25))
```

```python
cars = df[df['cylinders'] > 6]['car name'].tolist()
print(cars)
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

hp = df['horsepower'].to_numpy()

hp = np.where(np.isnan(hp), np.nanmean(hp), hp)

hp_updated = hp * 1.10

print(hp_updated[:10])
```

```python
disp = df['displacement'].to_numpy()
origin = df['origin'].to_numpy()

avg_disp = np.mean(disp[origin == 2])
print(avg_disp)
```

```python
cyl = df['cylinders'].to_numpy()
mpg = df['mpg'].to_numpy()

for c in np.unique(cyl):
    print(f"{c} cylinders →", np.mean(mpg[cyl == c]))
```

PANDAS

```python
print(df.head(10))
print("Shape:", df.shape)
print(df.describe())
```

```python
filtered = df[(df['model year'] == 75) & (df['weight'] < 3000)]

result = filtered[['car name', 'weight', 'mpg']]
print(result)
```

```python
print(df.isnull().sum())

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
```

```python
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
df.head()
```

```python
print(df.groupby('origin')['mpg'].mean())
```

```python
summary = df.groupby('model year')[['mpg','weight','horsepower']].mean()
print(summary)
```

```python
high_mpg = df[df['mpg'] > 30][['mpg','cylinders','horsepower','weight']]

high_mpg.to_csv('high_mpg_cars.csv', index=False)
```
